```python
class Human():
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:
            print(f"{self.name}は大人ではありません。")

human1 = Human("太郎",20)
human2 = Human("長太郎",21)
human3 = Human("小太郎",19)
list = [human1,human2,human3]

for i in range(0,len(list)):
    list[i].check_adult()

```
